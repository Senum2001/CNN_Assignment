```python
import torch
import torch.nn as nn

class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()
        self.convblock1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1),  # Added padding for better output size
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.convblock2 = nn.Sequential(
            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1),  # Added padding
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.convblock3 = nn.Sequential(
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1),  # Added padding
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.flatten = nn.Flatten()
        # The size after 3 convolution blocks with max pooling (each reduces size by half)
        # Input size: 128x128 -> After convblock1: 64x64 -> After convblock2: 32x32 -> After convblock3: 16x16
        self.classifier = nn.Sequential(
            nn.Linear(64 * 16 * 16, 128),  # Adjusted to match flattened output size
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 32),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(32, 9)  # 9 classes as per the dataset
        )
        
    def forward(self, x):
        x = self.convblock1(x)
        x = self.convblock2(x)
        x = self.convblock3(x)
        x = self.flatten(x)
        x = self.classifier(x)
        return x
```

```python
def calculate_accuracy(model: nn.Module, dataloader: torch.utils.data.DataLoader) -> tuple:
    correct = 0
    total = 0
    predictions = []
    with torch.no_grad():
        for data in dataloader:
            inputs, labels = data
            x, y = inputs.to(device), labels.to(device)
            outputs = model(x)
            _, predicted = torch.max(outputs, 1)
            predictions.append(predicted)
            total += y.size(0)
            correct += (predicted == y).sum().item()
    return 100 * correct / total, torch.cat(predictions)

# Unpack the returned values
train_accuracy, train_predictions = calculate_accuracy(model, train_dataloader)
test_accuracy, test_predictions = calculate_accuracy(model, test_dataloader)

# Print only the accuracy
print(f"Train accuracy: {train_accuracy:.2f}%")
print(f"Test accuracy: {test_accuracy:.2f}%")
```

```python
def train(model, train_loader, valid_loader, epochs, optimizer, criterion, device):
    """
    Train and validate the model.
    
    Args:
        model: The PyTorch model to train.
        train_loader: DataLoader for the training set.
        valid_loader: DataLoader for the validation set.
        epochs: Number of training epochs.
        optimizer: Optimizer for the model.
        criterion: Loss function.
        device: Device to run the training on (e.g., 'cuda' or 'cpu').
    
    Returns:
        train_accuracy_hist: List of training accuracies.
        test_accuracy_hist: List of validation accuracies.
        train_loss_hist: List of training losses.
        test_loss_hist: List of validation losses.
    """
    train_accuracy_hist = []
    test_accuracy_hist = []
    train_loss_hist = []
    test_loss_hist = []

    for epoch in range(epochs):
        model.train()  # Set model to training mode
        running_loss = 0.0
        correct_train = 0
        total_train = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            correct_train += (predicted == labels).sum().item()
            total_train += labels.size(0)

        train_accuracy = 100 * correct_train / total_train
        train_loss = running_loss / len(train_loader)
        train_accuracy_hist.append(train_accuracy)
        train_loss_hist.append(train_loss)

        # Validation step
        model.eval()  # Set model to evaluation mode
        running_loss = 0.0
        correct_valid = 0
        total_valid = 0

        with torch.no_grad():
            for images, labels in valid_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                running_loss += loss.item()
                _, predicted = outputs.max(1)
                correct_valid += (predicted == labels).sum().item()
                total_valid += labels.size(0)

        valid_accuracy = 100 * correct_valid / total_valid
        valid_loss = running_loss / len(valid_loader)
        test_accuracy_hist.append(valid_accuracy)
        test_loss_hist.append(valid_loss)

        print(f"Epoch [{epoch+1}/{epochs}], "
              f"Train Accuracy: {train_accuracy:.2f}%, "
              f"Valid Accuracy: {valid_accuracy:.2f}%, "
              f"Train Loss: {train_loss:.4f}, "
              f"Valid Loss: {valid_loss:.4f}")

    return train_accuracy_hist, test_accuracy_hist, train_loss_hist, test_loss_hist
```

```python
train_accuracy_hist, test_accuracy_hist, train_loss_hist, test_loss_hist = train(
    model=model,
    train_loader=train_loader,   
    valid_loader=valid_loader,   
    epochs=iterations,           
    optimizer=optimizer,
    criterion=loss,              
    device=device
)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_confusion_matrix(model, dataloader, classes, device):
    """
    Generate and plot a confusion matrix for the given model and dataloader.

    Args:
        model: The trained PyTorch model.
        dataloader: DataLoader for the test or validation dataset.
        classes: List of class names.
        device: Device on which the model and data are loaded.
    """
    model.eval()  # Set the model to evaluation mode
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Create confusion matrix
    cm = confusion_matrix(all_labels, all_preds, labels=range(len(classes)))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)

    # Plot confusion matrix
    fig, ax = plt.subplots(figsize=(8, 8))
    disp.plot(ax=ax, cmap='viridis', colorbar=True)
    plt.title('Confusion Matrix')
    plt.show()


classes = train_dataset.classes  # Class names from the training dataset
plot_confusion_matrix(model, test_loader, classes, device)
```
